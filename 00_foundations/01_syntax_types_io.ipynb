## Comments and the simplest program

- `#` starts a line comment.
- `print(...)` writes to stdout.
- Strings can use `"..."` or `'...'` — pick one and be consistent. Triple quotes `"""..."""` allow multiline.

````python
# This is a comment.
print("Hello, Python!")
````

## Variables and the core built-in types

Use `=` to bind a name to a value. Names should be `snake_case`. Use `type(x)` to see what a value is.

````python
name = "Dinakar"      # str
age = 30               # int
height_m = 1.78        # float
is_admin = True        # bool   (note: capital T/F)
nothing = None         # NoneType — Python's null

print(name, age, height_m, is_admin, nothing)
print(type(name), type(age), type(height_m), type(is_admin), type(nothing))
````

### Numbers

`int` is arbitrary precision (no overflow). `float` is IEEE 754 double — beware rounding.

````python
print(2 ** 100)           # int: huge, no overflow
print(7 / 2)              # 3.5   — / is true division (always float)
print(7 // 2)             # 3     — // is floor division
print(7 % 2)              # 1     — modulo (remainder)
print(0.1 + 0.2)          # 0.30000000000000004  — classic float gotcha
````

### Strings

Strings are *immutable* sequences of characters. Some essentials:

````python
s = "hello, world"
print(len(s))             # length
print(s.upper())          # methods return new strings; the original is unchanged
print(s[0], s[-1])        # index from front (0) and from back (-1)
print(s[7:12])            # slice [start:stop] — stop is exclusive
print("world" in s)       # membership test
````

### f-strings — the standard way to format

An `f"..."` string evaluates `{expr}` inside it. Use these everywhere; avoid `%`-formatting and `str.format()` in new code.

````python
name = "Dinakar"
score = 0.876543
print(f"{name} scored {score:.2%}")   # .2% → percent with 2 decimals → 87.65%
print(f"{name=}, {score=:.3f}")        # self-documenting form (3.8+)
````

## Type conversion

Conversions are *explicit*. Python does **not** silently turn `"3"` into `3`.

````python
x = int("42")             # str  -> int
y = float("3.14")         # str  -> float
z = str(99)               # int  -> str
b = bool("")              # empty string is falsy -> False
print(x, y, z, b)
````

### Truthiness

Anything *empty* or *zero* is **falsy**. Everything else is **truthy**. This shows up everywhere — `if some_list:` is idiomatic for "if not empty".

````python
for v in [0, 0.0, "", [], {}, None, False, 1, "x", [0]]:
    print(repr(v), "->", bool(v))
````
