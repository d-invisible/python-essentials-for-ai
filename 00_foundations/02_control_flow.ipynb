# 02 — Control Flow

How Python decides what to do next: `if`/`elif`/`else`, `for`, `while`, `break`, `continue`, and `match`.

Two rules that bite newcomers:
1. **Indentation defines blocks.** 4 spaces is the convention. Mixing tabs and spaces is an error.
2. **Colons end the header line** (`if cond:`, `for x in seq:`).

## `if` / `elif` / `else`

```python
score = 72

if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "F"

print(grade)
```

### Conditional expressions (ternary)

`a if cond else b` — useful for one-line assignments. Don't nest these.

```python
n = 7
parity = "even" if n % 2 == 0 else "odd"
print(parity)
```

### Boolean operators: `and`, `or`, `not`

`and` / `or` *short-circuit* and return the actual operand, not just `True`/`False`. Subtle but useful.

```python
print(True and "yes")     # "yes"  — returns the last truthy operand
print(False or "fallback") # "fallback" — returns first truthy operand
print(None or 0 or "" or "default")
```

## `for` loops — iterate over a sequence

Python's `for` is a *for-each*. You almost never write C-style index loops.

```python
for fruit in ["apple", "banana", "cherry"]:
    print(fruit.upper())
```

### `range()`, `enumerate()`, `zip()` — three you'll use forever

```python
# range(stop), range(start, stop), range(start, stop, step)
for i in range(3):
    print(i)
print("---")

# enumerate gives (index, value)
for i, fruit in enumerate(["apple", "banana", "cherry"]):
    print(i, fruit)
print("---")

# zip walks two (or more) sequences in parallel
names = ["Alice", "Bob", "Cleo"]
scores = [90, 75, 60]
for name, score in zip(names, scores, strict=True):
    print(f"{name}: {score}")
```

## `while` loops

Use when you don't know in advance how many iterations you need.

```python
n = 1
while n < 100:
    n *= 2
print(n)
```

### `break` and `continue`

- `break` — exit the *enclosing* loop immediately.
- `continue` — skip to the next iteration.

```python
for n in range(20):
    if n % 2 == 0:
        continue          # skip even numbers
    if n > 11:
        break             # stop completely once n exceeds 11
    print(n)
```

## `match` — structural pattern matching (3.10+)

Not just a `switch`. `match` can destructure tuples, dicts, and objects. Use it when you have several distinct shapes of input.

For simple `==` checks, prefer `if/elif`. `match` shines when you're pulling values *out* of a shape.

```python
def describe(point):
    match point:
        case (0, 0):
            return "origin"
        case (x, 0):
            return f"on x-axis at {x}"
        case (0, y):
            return f"on y-axis at {y}"
        case (x, y):
            return f"point ({x}, {y})"
        case _:
            return "not a 2D point"

for p in [(0, 0), (5, 0), (0, 3), (2, 4), "oops"]:
    print(p, "->", describe(p))
```

**Takeaways**

- Indentation = blocks. Use 4 spaces.
- `for x in seq:` is the everyday loop; `enumerate`/`zip`/`range` cover the rest.
