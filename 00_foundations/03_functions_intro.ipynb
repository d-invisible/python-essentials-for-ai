# 03 — Functions (intro)

Functions are how Python code is organized. This is the *intro* — we revisit `*args`/`**kwargs`, closures, and decorators in topic **02_functions_deep**.

**Anatomy:**
```python
def name(param1, param2=default) -> ReturnType:
    """One-line description."""
    ...
    return value
```

````python
def greet(name: str) -> str:
    """Return a friendly greeting for `name`."""
    return f"Hello, {name}!"

print(greet("Dinakar"))
print(greet.__doc__)        # docstrings are accessible at runtime
````

## Type hints

Python is dynamically typed, but **type hints** (`: str`, `-> int`) document intent and let editors/tools catch bugs. They are *not* enforced at runtime — Python won't raise if you pass the wrong type, but your IDE will warn you.

We'll use hints everywhere in this project. Get into the habit.

````python
def add(a: int, b: int) -> int:
    return a + b

print(add(2, 3))
print(add("foo", "bar"))    # works at runtime — type hints are not checks
````

## Default values and keyword arguments

- Parameters with `=` provide defaults; callers may omit them.
- Callers can pass arguments **positionally** or **by keyword** (`name=value`).
- **Never use a mutable default** like `[]` or `{}` — they are shared across calls. Use `None` and replace inside.

````python
def greet(name: str, greeting: str = "Hello") -> str:
    return f"{greeting}, {name}!"

print(greet("Dinakar"))
print(greet("Dinakar", "Hi"))
print(greet(name="Dinakar", greeting="Hey"))   # keyword form: order doesn't matter
````

````python
# THE MUTABLE-DEFAULT GOTCHA
def bad_append(item, bucket=[]):       # BAD: list created once, reused forever
    bucket.append(item)
    return bucket

print(bad_append(1))    # [1]
print(bad_append(2))    # [1, 2]  <- not what most people expect

def good_append(item, bucket=None):    # GOOD
    if bucket is None:
        bucket = []
    bucket.append(item)
    return bucket

print(good_append(1))   # [1]
print(good_append(2))   # [2]
````

## Return values

- A function without `return` returns `None`.
- You can return multiple values as a tuple, and unpack at the call site.

````python
def divmod_(a: int, b: int) -> tuple[int, int]:
    return a // b, a % b              # this is a tuple

quotient, remainder = divmod_(17, 5)  # tuple unpacking
print(quotient, remainder)
````

## Scope

- Variables defined inside a function are **local** to that call.
- Functions can *read* names from the enclosing module (global) scope.
- To *reassign* a global from inside a function, you need `global` — but reaching for it usually means your design is wrong. Prefer passing arguments and returning values.

````python
PI = 3.14159    # module-level ("global")

def area(r: float) -> float:
    # PI is read from the enclosing scope; no need to declare it.
    return PI * r * r

print(area(2))
````
