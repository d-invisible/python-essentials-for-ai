# 02 — Dicts and Sets

**`dict`** = key → value mapping. **`set`** = unique, unordered collection.

Both rely on **hashing**: keys/members must be *hashable* (immutable). So `str`, `int`, `tuple` work as keys; `list`, `dict`, `set` do not.

## Creating and accessing a dict

````python
user = {"name": "Dinakar", "age": 30, "admin": True}
print(user["name"])
user["age"] = 31              # update existing key
user["email"] = "x@y.z"      # add new key
del user["admin"]              # remove a key
print(user)
print("name" in user)          # membership test = key lookup, O(1) avg
````

### Missing keys — use `get`, not `[]`

`d[k]` raises `KeyError` if `k` is missing. `d.get(k, default)` returns the default instead. Use `[]` when the key *must* exist (and you want the loud failure), `get` when it might not.

````python
print(user.get("name"))
print(user.get("phone"))               # None
print(user.get("phone", "unknown"))  # "unknown"

try:
    user["phone"]
except KeyError as e:
    print("missing:", e)
````

### Iterating a dict

- `for k in d:` → iterates keys.
- `for k, v in d.items():` → keys and values together. **Use this.**
- `d.values()` if you only want values.

Insertion order is preserved (since 3.7) — iteration order = insertion order.

````python
scores = {"alice": 90, "bob": 75, "cleo": 60}
for name, score in scores.items():
    print(f"{name}: {score}")
````

## Common dict patterns

### Counting

````python
text = "the quick brown fox jumps over the lazy dog the end"

# manual
counts = {}
for word in text.split():
    counts[word] = counts.get(word, 0) + 1
print(counts)

# idiomatic: collections.Counter
from collections import Counter
print(Counter(text.split()))
print(Counter(text.split()).most_common(3))
````

### Grouping — `setdefault` or `defaultdict`

````python
from collections import defaultdict

people = [
    ("engineering", "Alice"),
    ("engineering", "Bob"),
    ("sales", "Cleo"),
    ("sales", "Dan"),
    ("engineering", "Eve"),
]

# setdefault (built-in)
groups = {}
for dept, name in people:
    groups.setdefault(dept, []).append(name)
print(groups)

# defaultdict (cleaner)
groups2 = defaultdict(list)
for dept, name in people:
    groups2[dept].append(name)
print(dict(groups2))   # cast back to plain dict for printing
````

### Merging dicts

````python
defaults = {"theme": "light", "timeout": 30}
overrides = {"timeout": 60, "debug": True}

# 3.9+ — merge operator. Later keys win.
merged = defaults | overrides
print(merged)

# in-place merge:
config = defaults.copy()
config |= overrides
print(config)
````

## Sets — fast membership, no duplicates

````python
tags = {"python", "web", "async", "python"}
print(tags)                    # duplicates collapsed
print("python" in tags)        # O(1) avg
tags.add("fastapi")
tags.discard("web")            # discard = remove if present, no error if not
print(tags)

# {} is an empty DICT, not an empty set. Use set() for an empty set.
empty = set()
print(type(empty), empty)
````

### Set algebra

````python
a = {1, 2, 3, 4}
b = {3, 4, 5, 6}

print("union       :", a | b)
print("intersection:", a & b)
print("difference  :", a - b)        # in a but not b
print("sym. diff   :", a ^ b)        # in exactly one
print("subset      :", {1, 2} <= a)
````

### Dedup while preserving order

A plain `set` discards order. To dedup *and* keep first-seen order, use a `dict` — Python guarantees insertion order on dicts:

````python
items = ["b", "a", "a", "c", "b", "d"]
deduped = list(dict.fromkeys(items))   # neat trick
print(deduped)
````

**Takeaways**

- Use `d.get(k, default)` when a key may be missing; `d[k]` when it must exist.
- `Counter` and `defaultdict` (from `collections`) replace 90% of manual counting/grouping code.
- Sets give O(1) membership and clean algebra (`|`, `&`, `-`, `^`).
- Keys/members must be hashable — `list` and `dict` can't be either.
